```python
import pickle

with open("../data/stats.pkl", "rb") as f:
        stats_list = pickle.load(f)
```

```python
print(len(stats_list))
```

```python
import matplotlib.pyplot as plt

data = stats_list

combined_len = [d['combined_len'] for d in data]

plt.plot(range(len(combined_len)), combined_len, marker='o', linestyle='-', color='blue')
plt.xlabel('Sample Index')
plt.ylabel('Combined Length')
plt.title('Combined Length Curve over Samples')
plt.grid(True)
plt.show()

duration_seconds = [d['duration_seconds'] for d in data]

plt.plot(range(len(duration_seconds)), duration_seconds, marker='o', linestyle='-', color='blue')
plt.xlabel('Sample Index')
plt.ylabel('Duration seconds')
plt.title('Combined Length Curve over Samples')
plt.grid(True)
plt.show()
```
